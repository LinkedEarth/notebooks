# Make composites

In this notebook, we demonstrate how to make composites from datasets in the PAGES 2k compilation, grouping them by archive. We limit the analysis to corals, glacier ice, lake & marine sediments, as well as wood, so that there are enough series to make meaningful composites. 
This emulates Fig 8 of [PAGES 2k Consortium (2017)](https://www.nature.com/articles/sdata201788) (minus documentary records). 

```python
import pandas as pd
import p2k
```

```python
# load the PAGES2k dataset
df = pd.read_pickle('./data/pages2k_dataset.pkl')
df = p2k.clean_df(df)
```

After checking, observation types that have negative correlation to temperature:
+ Coral - calcification
+ Coral - d18O
+ Coral - Sr/Ca

We will revert the sign of those records first, and then stratify records by archive types and make a composite for each type.

```python
for index, row in df.iterrows():
    if row['archiveType'] == 'coral' and row['paleoData_variableName'] in ['calcification', 'd18O', 'Sr_Ca']:
        row['paleoData_values'] *= -1
```

Please refer to the `p2k` package for the details of the function `make_composite()`, which calls a couple of other functions in that package.
Essentially, it calls:
+ `df_append_nearest_obs()`, which find the nearest instrumental observation value for each proxy location and append that value to the dataframe of proxy records
+ `df2composite()`, which performs binning and averaging of the proxy timeseries, and bootstrap for uncertainty estimation
+ `df2comp_ols()`, which calculates the linear regression statistics
+ `plot_composite()`, which plot the results

```python
HadCRUT4_path = './data/HadCRUT.4.6.0.0.median.nc'
archive_types = [
    'coral',
    'glacier ice',
    'lake sediment',
    'marine sediment',
    'tree',
]

bin_width = 10
ys_proxy = {}  # the value-axis of the composite
ts_proxy = {}  # the time-axis of the composite
slope = {}  # the calculated slope of the linear regression
R2 = {}  # the squared correlation of the linear regression
for archive_type in archive_types:
    print(f'Processing {archive_type}...')
    selector = df['archiveType']==archive_type
    df_archive = df[selector]
    ys_proxy[archive_type], ts_proxy[archive_type], slope[archive_type], R2[archive_type], fig = p2k.make_composite(
        df_archive, inst_temp_path=HadCRUT4_path, archive_type=archive_type, bin_width=bin_width,
        title=f'{archive_type}, {len(df_archive)} records, bin_width={bin_width}'
    )
    fig.savefig(f'./figs/pages2k_composite_{archive_type}.png', bbox_inches='tight')
```
